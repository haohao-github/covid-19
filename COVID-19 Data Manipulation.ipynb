# Transition Data

```python
import pandas as pd
import re
from datetime import datetime
```

```python
## take a look at one of the data files
url = 'https://github.com/COVIDExposureIndices/COVIDExposureIndices/blob/master/lex_data/state_lex_2020-01-20.csv?raw=true'
df = pd.read_csv(url, header = 0)
df.head()
```

```python
## create date list
dates = pd.date_range(start = "2020-01-20",end = "2020-04-06")
date_list = []
for date in dates:
    date = re.findall(r'2020-\d{2}-\d{2}', str(date))
    date_list.append(date[0])
date_list
```

```python
## read and combine all the data files
transition = pd.DataFrame(columns = ['STATE_PRE', 'STATE', 'LEX', 'DATE'])

for date in date_list:
    url = 'https://github.com/COVIDExposureIndices/COVIDExposureIndices/blob/master/lex_data/state_lex_' + date + '.csv?raw=true'
    df = pd.read_csv(url, header=0)
    df = pd.melt(df, id_vars="STATE_PRE", var_name = "STATE", value_name = "LEX")
    df['DATE'] = date
    transition = transition.append(df)

transition.head()
```

```python
## save as a csv file
transition.to_csv(r'transition.csv',index = False)
```

# COVID-19 Data

### confirmed data
https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset

```python
## read confirmed data
confirmed = pd.read_csv('time_series_covid_19_confirmed_US.csv', header = 0)
confirmed.head()
```

```python
## create column list for filtering
col_list = [6]
for i in range(11, len(confirmed.columns)):
    col_list.append(i)
```

```python
## filter columns
confirmed = confirmed.iloc[:,col_list]
confirmed.head()
```

```python
## read state code reference table
state_abbr = pd.read_csv('state_abbr.csv', header=0)
state_abbr.head()
```

```python
## rename and filter columns
state_abbr = state_abbr.rename(columns = {'State': 'Province_State'})
state_abbr = state_abbr.iloc[:, [0,2]]
state_abbr.head()
```

```python
## add state code
confirmed_result = pd.merge(confirmed, state_abbr, on = 'Province_State', how = 'left')
```

```python
## gather the data
confirmed_result = pd.melt(confirmed_result.iloc[:,1:len(confirmed_result)], id_vars='Code', var_name = 'date', value_name = 'confirmed')
```

```python
## sum up the number of confirmed cases by state
sum_confirmed = confirmed_result.groupby(['Code','date'], as_index=False).sum()
sum_confirmed.head()
```

### deaths data
https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset

```python
## read death data
deaths = pd.read_csv('time_series_covid_19_deaths_US.csv', header = 0)
deaths.head()
```

```python
## create column list for filtering
col_list_d = [6]
for i in range(11, len(deaths.columns)):
    col_list_d.append(i)
```

```python
## filter the columns
deaths = deaths.iloc[:,col_list_d]
deaths.head()
```

```python
## add state code
deaths_result = pd.merge(deaths, state_abbr, on='Province_State', how = 'left')
```

```python
## gather the data
deaths_result = pd.melt(deaths_result.iloc[:,1:len(deaths_result)], id_vars=['Code','Population'], var_name = 'date', value_name = 'deaths')
```

```python
## sum up the number of population and deaths by state
sum_deaths = deaths_result.groupby(['Code','date'], as_index=False).sum()
sum_deaths.head()
```

### Combine confirmed and death data

```python
## combine confirmed and death
covid_data = pd.merge(sum_deaths, sum_confirmed, on=['Code', 'date'], how = 'inner')
covid_data = covid_data.rename(columns = {'Code': 'state'})
covid_data.head()
```

```python
## transform date format
for i in range(len(covid_data)):
    date = datetime.strptime(covid_data.iloc[i,1], '%m/%d/%y')
    covid_data.iloc[i,1] = date.strftime("%m/%d/%y")
```

```python
## save as a csv file
covid_data.to_csv(r'covid_data.csv',index = False)
```

## Daily Data (including test numbers)

```python
url = 'https://covidtracking.com/api/us/daily.csv'
dd = pd.read_csv(url, header = 0)
dd.head()
```

```python
## filter and rename the columns
dd = dd[['date', 'states', 'deathIncrease', 'hospitalizedIncrease', 'positiveIncrease', 'totalTestResultsIncrease', 'death', 'hospitalized', 'positive', 'totalTestResults']]
dd = dd.rename(columns = {'deathIncrease': 'deaths', 'hospitalizedIncrease': 'hospitalized', 'positiveIncrease': 'positive', 'totalTestResultsIncrease': 'tests', 'death': 'death_cum', 'hospitalized': 'hospitalized_cum', 'positive': 'positive_cum', 'totalTestResults': 'tests_cum'})
dd.head()
```

```python
## fill na with 0
for col in dd.columns:
    dd[col].fillna(value = 0, inplace = True)
```

```python
## convert some negative 'tests' numbers to positive
for row in range(len(dd)):
    dd.iloc[row, 5] = abs(dd.iloc[row, 5])
dd.head()
```

```python
## save as a csv file
dd.to_csv(r'daily_data.csv',index = False)
```
